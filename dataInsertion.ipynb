### Insertion of First Dataset - [ Vaccination Details Worldwide ]

```python
covidVaccination = pd.read_csv(r"C:\Users\Sabharish\OneDrive\Desktop\inProgress\semProject\BDMS\data\country_vaccinations.csv")
covidVaccination.head()
```

```python
columns = ['country', 'iso_code', 'date', 'vaccines', 'source_name', 'source_website', 'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated',
           'daily_vaccinations_raw', 'daily_vaccinations', 
           'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred', 
           'people_fully_vaccinated_per_hundred', 'daily_vaccinations_per_million']

print(len(columns))
```

```python
db.vaccinations.delete_many({})
for i in range(len(covidVaccination)):
    try:
        array = {}
        for k in range(len(columns)):
            if covidVaccination[columns[k]][i] is not None and str(covidVaccination[columns[k]][i]).strip() not in ['', None, 'nan', 'NaN']:
                 array[columns[k]] = covidVaccination[columns[k]][i]                
        col.insert_one(array)
    except Exception as e:
        print(e)
        continue

print("Finished")
```
